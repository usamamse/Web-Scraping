```python
!pip install pandas
import pandas as pd
from bs4 import BeautifulSoup
import json
import requests
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

url = 'https://understat.com/league/EPL/2023'
url2 = 'https://understat.com/league/EPL/2022'
url3 = 'https://understat.com/league/EPL/2021'
url4 = 'https://understat.com/league/EPL/2020'

res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')

strings = scripts[2].string

ind_start = strings.index("('")+2
ind_end = strings.index("')")
json_data = strings[ind_start:ind_end]
json_data = json_data.encode("utf8").decode("unicode_escape")
data = json.loads(json_data)
#=============================================

res = requests.get(url2)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')

strings2 = scripts[2].string

ind_start = strings2.index("('")+2
ind_end = strings2.index("')")
json_data = strings2[ind_start:ind_end]
json_data = json_data.encode("utf8").decode("unicode_escape")
data2 = json.loads(json_data)
#=============================================

res = requests.get(url3)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')

strings3 = scripts[2].string

ind_start = strings3.index("('")+2
ind_end = strings3.index("')")

json_data = strings3[ind_start:ind_end]
json_data = json_data.encode("utf8").decode("unicode_escape")
data3 = json.loads(json_data)
#=============================================

res = requests.get(url4)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')

strings4 = scripts[2].string

ind_start = strings4.index("('")+2
ind_end = strings4.index("')")
json_data = strings4[ind_start:ind_end]
json_data = json_data.encode("utf8").decode("unicode_escape")
data4 = json.loads(json_data)


#print(json.dumps(data, indent=2))

start_date = datetime.strptime('2023-08-09', '%Y-%m-%d')
end_date = datetime.strptime('2023-08-16', '%Y-%m-%d')

xGA_goals_data = []
for team_id, team_data in data.items():
    team_name = team_data['title']
    if 'history' in team_data:
        for game in team_data['history']:
            game_date = datetime.strptime(game['date'], '%Y-%m-%d %H:%M:%S')

            if start_date <= game_date <= end_date:
              if team_name == 'Luton' or team_name == 'Sheffield United' or team_name == 'Burnley':
                    xGA = game.get('xGA', 0)
                    goalsagainst = game.get('missed', 0)
                    xGA_goals_data.append({'team': team_name, 'season': '2023/24', 'xGA': xGA, 'goals against': goalsagainst}) 

df = pd.DataFrame(xGA_goals_data)

#======================================================================

start_date2 = datetime.strptime('2022-08-04', '%Y-%m-%d')
end_date2 = datetime.strptime('2022-08-08', '%Y-%m-%d')

xGA_goals_data2 = []
for team_id, team_data in data2.items():
    team_name = team_data['title']
    if 'history' in team_data:
        for game in team_data['history']:
            game_date = datetime.strptime(game['date'], '%Y-%m-%d %H:%M:%S')

            if start_date2 <= game_date <= end_date2:
                if team_name == 'Fulham' or team_name == 'Bournemouth' or team_name == 'Nottingham Forest': 
                    xGA = game.get('xGA', 0)
                    goalsagainst = game.get('missed', 0)
                    xGA_goals_data2.append({'team': team_name, 'season': '2022/23', 'xGA': xGA, 'goals against': goalsagainst}) 

df2 = pd.DataFrame(xGA_goals_data2)

#======================================================================

start_date3 = datetime.strptime('2021-08-12', '%Y-%m-%d')
end_date3 = datetime.strptime('2021-08-16', '%Y-%m-%d')

xGA_goals_data3 = []
for team_id, team_data in data3.items():
    team_name = team_data['title']
    if 'history' in team_data:
        for game in team_data['history']:
            game_date = datetime.strptime(game['date'], '%Y-%m-%d %H:%M:%S')

            if start_date3 <= game_date <= end_date3:
                if team_name == 'Norwich' or team_name == 'Watford' or team_name == 'Brentford':
                    xGA = game.get('xGA', 0)
                    goalsagainst = game.get('missed', 0)
                    xGA_goals_data3.append({'team': team_name, 'season': '2021/22', 'xGA': xGA, 'goals against': goalsagainst}) 

df3 = pd.DataFrame(xGA_goals_data3)

#======================================================================

start_date4 = datetime.strptime('2020-09-11', '%Y-%m-%d')
end_date4 = datetime.strptime('2020-09-15', '%Y-%m-%d')

xGA_goals_data4 = []
for team_id, team_data in data4.items():
    team_name = team_data['title']
    if 'history' in team_data:
        for game in team_data['history']:
            game_date = datetime.strptime(game['date'], '%Y-%m-%d %H:%M:%S')

            if start_date4 <= game_date <= end_date4:
                if team_name == 'Leeds' or team_name == 'West Bromwich Albion' or team_name == 'Fulham':
                    xGA = game.get('xGA', 0)
                    goalsagainst = game.get('missed', 0)
                    xGA_goals_data4.append({'team': team_name, 'season': '2020/21', 'xGA': xGA, 'goals against': goalsagainst}) 

df4 = pd.DataFrame(xGA_goals_data4)

#======================================================================

df_melted = df.melt(id_vars=['team'], value_vars=['xGA','goals against'], var_name='metric',value_name='value')
df2_melted = df2.melt(id_vars=['team'], value_vars=['xGA','goals against'], var_name='metric',value_name='value')
df3_melted = df3.melt(id_vars=['team'], value_vars=['xGA','goals against'], var_name='metric',value_name='value')
df4_melted = df4.melt(id_vars=['team'], value_vars=['xGA','goals against'], var_name='metric',value_name='value')

fig, axes = plt.subplots(2,2, figsize=(18, 18), sharey= True)
sns.barplot(data = df_melted, x= 'team', y='value', hue='metric', ax=axes[0,0])
axes[0,0].set_title('Gameweek 1 - 2023/24 (Promoted Teams)', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel("Teams", fontsize=12)
axes[0,0].set_ylabel("Goals Against/xGA")
axes[0,0].set_ylim(0,5)
axes[0,0].grid(True)

sns.barplot(data = df2_melted, x= 'team', y='value', hue='metric', ax=axes[0,1])
axes[0,1].set_title('Gameweek 1 - 2022/23 (Promoted Teams)', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel("Teams", fontsize=12)
axes[0,1].set_ylabel("Goals Against/xGA")
axes[0,1].set_ylim(0,5)
axes[0,1].grid(True)

sns.barplot(data = df3_melted, x= 'team', y='value', hue='metric', ax=axes[1,0])
axes[1,0].set_title('Gameweek 1 - 2021/22 (Promoted Teams)', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel("Teams", fontsize=12)
axes[1,0].set_ylabel("Goals Against/xGA")
axes[1,0].set_ylim(0,5)
axes[1,0].grid(True)

sns.barplot(data = df4_melted, x= 'team', y='value', hue='metric', ax=axes[1,1])
axes[1,1].set_title('Gameweek 1 - 2020/21 (Promoted Teams)', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel("Teams", fontsize=12)
axes[1,1].set_ylabel("Goals Against/xGA")
axes[1,1].set_ylim(0,5)
axes[1,1].grid(True)

plt.savefig('Gameweek 1 xGC Promoted Teams.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
